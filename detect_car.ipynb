```python
from ultralytics import YOLO
import cv2
```

```python
model= YOLO("yolov8n.pt")
```

```python
import torch

# Check if CUDA is available
if torch.cuda.is_available():
    device = torch.device("cuda")
else:
    device = torch.device("cpu")
print(device)
```

```python
# input image
image= "AWS Images\image_017.jpg"
```

```python
results = model.predict(image,save=True,device=device,save_txt=True, conf=0.5,project='Output', name='img', classes=[2],max_det=2)

for result in results:
    out=result.save_dir

print(out)
```

```python
import csv
import os

labels_folder = str(out) + '\labels'
print(labels_folder)

if not os.listdir(labels_folder):
    output_value = 0
else:
    output_value = 1

output_csv_path = "output2.csv"
fieldnames = ['ImageName', 'Is_car_present']  # Corrected fieldnames

# Check if the CSV file already exists
csv_exists = os.path.isfile(output_csv_path)

with open(output_csv_path, 'a', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    # If the CSV file doesn't exist, write the header
    if not csv_exists:
        writer.writeheader()

    # Write the data
    writer.writerow({'ImageName': image, 'Is_car_present': output_value})

print(f'Is_car_present: {output_value}')
```
